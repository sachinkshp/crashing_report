```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming the data is in a pandas DataFrame called `data`
filtered_data = data[(data['Speed Limit'] >= 15) & (data['Speed Limit'] <= 50)]

plt.figure(figsize=(10,10))
plt.title("Speed Limit")
plt.boxplot(filtered_data['Speed Limit'])
plt.show()
```

```python
# Create a list of categories to include in the plot
categories = ['NO CONTROLS', 'Country', 'STOP SIGN', 'FLASHING TRAFFIC SIGNAL']

# Filter the data based on the categories list
filtered_data = data[data['Traffic Control'].isin(categories)]

# Create the countplot using the filtered data
plt.figure(figsize=(15,7))
plt.title("Traffic signal on roads")
order = filtered_data['Traffic Control'].value_counts(ascending=False).index 
sns.countplot(x='Traffic Control', data=filtered_data, order=order)

# Set the rotation angle of x-axis labels to 45 degrees
plt.xticks(rotation=45)

plt.show()
```

```python
# Create a list of categories to include in the plot
categories = ['Maryland (State)', 'County', 'Municipality', 'US (State)']

# Filter the data based on the categories list
filtered_data = data[data['Route Type'].isin(categories)]

# Create the countplot using the filtered data
plt.figure(figsize=(15,7))
plt.title("Proportion of accidens on route type")
order = filtered_data['Route Type'].value_counts(ascending=False).index 
sns.countplot(x='Route Type', data=filtered_data, order=order)

# Set the rotation angle of x-axis labels to 45 degrees
plt.xticks(rotation=45)

plt.show()
```

```python
# Create a list of categories to include in the plot
categories = ['Maryland (State)', 'County', 'Municipality', 'US (State)']

# Filter the data based on the categories list
filtered_data = data[data['Route Type'].isin(categories)]

# Create the countplot using the filtered data
plt.figure(figsize=(15,7))
plt.title("Proportion of accidens on route type")
order = filtered_data['Light'].value_counts(ascending=False).index 
sns.countplot(x='Light', data=filtered_data, order=order)

# Set the rotation angle of x-axis labels to 45 degrees
plt.xticks(rotation=45)

plt.show()
```

```python

# Get the count of accidents for each type of light condition
accident_counts = filtered_data['Light'].value_counts(ascending=False)

# Sort the counts by index to ensure they are plotted in the correct order
accident_counts = accident_counts.sort_index()

# Create the line plot
sns.lineplot(x=accident_counts.index, y=accident_counts.values)

# Set the rotation angle of x-axis labels to 45 degrees
plt.xticks(rotation=45)

# Add title to the plot
plt.title("Causes of accidents")

plt.show()
```

```python
sns.set_style("white")
corr = data.corr().round(2)

mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

f, ax = plt.subplots(figsize=(10, 10))

cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap
sns.heatmap(corr, mask=mask, cmap=cmap, vmin=-1, vmax=1, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5}, annot=True)

plt.tight_layout()
```
